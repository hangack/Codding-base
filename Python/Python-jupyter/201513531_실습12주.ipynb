```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from mpl_toolkits.basemap import Basemap
from matplotlib import cm
```

```python
plt.figure(figsize=(10,10))

map = Basemap(projection='ortho', lat_0=37.35, lon_0=126.58, resolution='i', area_thresh=1000)
### basemap  
### https://matplotlib.org/basemap/api/basemap_api.html
### https://matplotlib.org/basemap/users/geography.html
map.drawcoastlines() ##  
map.drawcountries()  ##  
map.fillcontinents(color='skyblue') ##   
map.drawmapboundary()
map.drawmeridians(np.arange(0,360,30))  ##     (,,)
map.drawparallels(np.arange(-90,90,30)) ## 

plt.show()      ##   
```

```python
plt.figure(figsize=(10,10))

map = Basemap(projection='merc', lat_0=37.35, lon_0=126.58, resolution='h', urcrnrlat=44, llcrnrlat=32, urcrnrlon=132.5, llcrnrlon=121.5)
map.drawcoastlines() 
map.drawcountries() 

plt.show()
```

```python
plt.figure(figsize=(10,10))

## ll:lower lest / ur:upper right / y:lat / x:lon
map = Basemap(projection='merc', lat_0=37.35, lon_0=126.58, resolution='h', urcrnrlat=44, llcrnrlat=32, urcrnrlon=132.5, llcrnrlon=121.5)
map.drawcoastlines() 
map.drawcountries() 

lon = 126.58
lat = 37.35
x,y = map(lon,lat)
map.plot(x, y, 'bo', markersize=12)  ## matplot

plt.show()
```

```python
x1 = -20
x2 = 40
y1 = 32
y2 = 64

m = Basemap(projection='merc',resolution='i',\
           llcrnrlat=y1, urcrnrlat=y2, llcrnrlon=x1, urcrnrlon=x2)       ##  \ :    ## lat_ts : defalt=0 ()
```

```python
plt.figure(figsize=(8,7))
m.drawcoastlines(linewidth=1.0)

plt.show()
```

```python
plt.figure(figsize=(8,7))
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='gray',lake_color='aqua', alpha=0.5)  ## alpha :  0 ~ 1
#m.drawlsmask(land_color='gray',ocean_color='aqua')
plt.show()
```

```python
plt.figure(figsize=(8,7))
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='gray',lake_color='aqua', alpha=0.5)
m.etopo()       ##  
plt.show()
```

```python
plt.figure(figsize=(8,7))
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='gray',lake_color='aqua', alpha=0.5)
m.bluemarble()
plt.show()
```

```python
plt.figure(figsize=(8,7))
m.drawcoastlines()
m.drawcountries()
m.fillcontinents(color='gray',lake_color='aqua', alpha=0.5)
m.shadedrelief()
plt.show()
```
