```python
# 질문 리스트 구하기
part_questions = []
part_questions = ['Q' + str(7) + '_Part_' + str(j) for j in range(1, 13)]
part_questions.append('Q' + str(7) + '_OTHER')
part_questions

# category 별 전체 카운트 구하기
programmings = []
counts = []
for i in part_questions:
  programming = df[i].value_counts().index[0]
  val = df[i].value_counts()[0]
  programmings.append(programming)
  counts.append(val)

combined_df = pd.DataFrame()
combined_df['Programming'] = programmings
combined_df['Count'] = counts

combined_df = combined_df.sort_values(['Count'], ascending = False)
combined_df
```
